```python

def prob_calcs(a, w, i, j):
  probs = []
  if a == 'r':
    if i == 0 and j == 9:
      probs.append(1-w+(w/4)+(w/4)) #r
      probs.append(w/4) #l
      probs.append(0) #u
      probs.append(w/4) #d
    elif i == 9 and j == 9:
      probs.append(1-w+(w/4) + (w/4))
      probs.append(w/4)
      probs.append(w/4)
      probs.append(0)
    elif i == 0 and j == 0:
      probs.append(1-w+(w/4))
      probs.append((w/4)+(w/4))
      probs.append(0)
      probs.append((w/4))
    elif i == 9 and j == 0:
      probs.append(1-w+(w/4))
      probs.append((w/4)+(w/4))
      probs.append((w/4))
      probs.append(0)
    else:
      probs.append(1-w+(w/4))
      probs.append(w/4)
      probs.append(w/4)
      probs.append(w/4)
  elif a == 'l':
    if i == 0 and j == 0:
      probs.append(w/4)
      probs.append(1-w+(w/4)+(w/4))
      probs.append(0)
      probs.append(w/4)
    elif i == 9 and j == 0:
      probs.append(w/4)
      probs.append(1-w+(w/4)+(w/4))
      probs.append(w/4)
      probs.append(0)
    elif i == 0 and j == 9:
      probs.append((w/4)+(w/4)) #r
      probs.append(1-w+(w/4)) #l
      probs.append(0) #u
      probs.append(w/4) #d
    elif i == 9 and j == 9:
      probs.append((w/4)+(w/4))
      probs.append(1-w+(w/4))
      probs.append(w/4)
      probs.append(0)
    else:
      probs.append(w/4)
      probs.append(1-w+(w/4))
      probs.append(w/4)
      probs.append(w/4)
  elif a == 'u':
    if i == 0 and j == 0:
      probs.append(w/4)
      probs.append(0)
      probs.append(1-w+(w/4)+(w/4))
      probs.append(w/4)
    elif i == 0 and j == 9:
      probs.append(0)
      probs.append(w/4)
      probs.append(1-w+(w/4)+(w/4))
      probs.append(w/4)
    elif i == 9 and j == 0:
      probs.append(w/4)
      probs.append((w/4)+(w/4))
      probs.append(1-w+(w/4))
      probs.append(0)
    elif i == 9 and j == 9:
      probs.append((w/4)+(w/4))
      probs.append(w/4)
      probs.append(1-w+(w/4))
      probs.append(0)
    else:
      probs.append(w/4)
      probs.append(w/4)
      probs.append(1-w+(w/4))
      probs.append(w/4)
  else:
    if i == 9 and j == 9:
      probs.append(0)
      probs.append(w/4)
      probs.append(w/4)
      probs.append(1-w+(w/4)+(w/4))
    elif i == 9 and j == 0:
      probs.append(w/4)
      probs.append(0)
      probs.append(w/4)
      probs.append(1-w+(w/4)+(w/4))
    elif i == 0 and j == 0:
      probs.append(w/4)
      probs.append((w/4)+(w/4))
      probs.append(0)
      probs.append(1-w+(w/4))
    elif i == 0 and j == 9:
      probs.append((w/4)+(w/4))
      probs.append(w/4)
      probs.append(0)
      probs.append(1-w+(w/4))
    else:
      probs.append(w/4)
      probs.append(w/4)
      probs.append(w/4)
      probs.append(1-w+(w/4))
  return probs
```

```python
def value_calc(i, j, V, a, g, w, reward_f):
  #prob * (reward + (gamma * value))
  reward = 0
  probabilities = prob_calcs(a, w, i, j)
  r_j = j + 1
  l_j = j - 1
  u_i = i - 1
  d_i = i+1

  #move right
  if r_j == 10:  #off grid
    #print("move right off grid ", reward)
    reward += probabilities[0] * (reward_f[i][j] + (g*V[i][j]))
  else: #on grid so continue
    #print("move right ", reward)
    reward += probabilities[0] * (reward_f[i][r_j] + (g*V[i][r_j]))
  #move left
  if l_j == -1:  #off grid
    #print("move left off grid ", reward)
    reward += probabilities[1] * (reward_f[i][j] + (g*V[i][j]))
  else:  #on grid so continue
    #print("move left ", reward)
    reward += probabilities[1] * (reward_f[i][l_j] + (g*V[i][l_j]))
  #move up
  if u_i == -1: #off grid
    #print("move up off grid ", reward)
    reward += probabilities[2] * (reward_f[i][j] + (g*V[i][j]))
  else: #on grid so continue
    #print("move up ", reward)
    reward += probabilities[2] * (reward_f[u_i][j] + (g*V[u_i][j]))
  #move down
  if d_i == 10: #off grid
    #print("move down off grid ", reward)
    reward += probabilities[3] * (reward_f[i][j] + (g*V[i][j]))
  else: #on grid so continue
    #print("move down ", reward)
    reward += probabilities[3] * (reward_f[d_i][j] + (g*V[d_i][j]))
  #print(reward)
  return reward
```

```python
def optimal(S, g, e, R, w):
  V_history = []
  V = np.zeros((10, 10))
  policy = np.zeros((10,10))
  delta = np.inf
  step_count = 0

  while delta > e:
    delta = 0
    for s in range(len(S)):
      for s_sub in range(len(S[s])):
        i = s
        j = s_sub
        v = V[s][s_sub]
        temp = (value_calc(i, j, V, 'r', g, w, R), value_calc(i, j, V, 'l', g, w, R), value_calc(i, j, V, 'u', g, w, R), value_calc(i, j, V, 'd', g, w, R))
        policy[s][s_sub] = np.argmax(temp)
        V[s][s_sub] = max(temp)
        delta = max(delta, abs(v - V[s][s_sub]))
    if step_count % 5 == 0:
      V_history.append(V.copy())

    step_count += 1

  return V, V_history, step_count, policy
```

```python
#10
def make_matrix(probs, a, i, j): #convert probability list into transition matrix
  mm = np.zeros((10,10))
  if j == 9:
    r_j = j
  else:
    r_j = j+1
  if j == 0:
    l_j = j
  else:
    l_j = j-1
  if i == 0:
    u_i = i
  else:
    u_i = i -1
  if i == 9:
    d_i = i
  else:
    d_i = i+1
  mm[i][r_j] += probs[0]
  mm[i][l_j] += probs[1]
  mm[u_i][j] += probs[2]
  mm[d_i][j] += probs[3]

  return mm


```

```python
def lp_matrices(lmbd, a_policy, r_max):
  p_a1 = []
  p_a = []
  p_a.append([])
  p_a.append([])
  p_a.append([])

  c = np.concatenate([np.ones(100), -lmbd*np.ones(100), np.zeros(100)])
  c = -c

  I = np.identity(100)

  Z = np.zeros((100,100), dtype = float)#.reshape(100, 100)

  D = np.concatenate((Z, Z, Z, Z))
  D = np.concatenate((D, np.concatenate((-I, -I, Z, Z)), np.concatenate((I, -I, I, -I))), 1)

  D1 = []
  D2 = []
  for s in range(100):
    i = s // 10
    j = s % 10
    p_move = a_policy[i][j]
    idx = 0
    for k in range(4):
      if k == 0:
        move = 'r'
      elif k == 1:
        move = 'l'
      elif k == 2:
        move = 'u'
      elif k == 3:
        move = 'd'
      temp_probs = prob_calcs(move, w, i, j)
      mm = make_matrix(temp_probs, move, i, j)
      if k == p_move:
        p_a1.append(mm.flatten())
      else:
        p_a[idx].append(mm.flatten())
        idx += 1

  for a in p_a:
    temp = np.dot(-(np.array(p_a1) - np.array(a)), np.linalg.inv(I - g*np.array(p_a1)))

    temp2 = np.concatenate((I, Z, temp), axis=1)
    temp3 = np.concatenate((Z, Z, temp), axis=1)

    if len(D1) == 0:
      D1 = temp2
    else:
      D1 = np.concatenate((D1, temp2))
    if len(D2) == 0:
      D2 = temp3
    else:
      D2 = np.concatenate((D2, temp3))

  D = np.concatenate((D1, D2, D))

  temp5 = np.zeros(800)
  temp6 = r_max*np.ones(200)

  b = np.concatenate((temp5, temp6))

  c = matrix(c)
  D = matrix(D)
  b = matrix(b)

  return c, D, b
```

```python
def solve_irl(c, D, b):
  solvers.options['show_progress'] = False
  sol = solvers.lp(c=c, G=D, h=b)
  return sol['x'][-100:]
```

```python
def accuracy(exp_pol, ag_pol, S_len):
  count = 0
  for i in range(len(exp_pol)):
    for j in range(len(exp_pol[i])):

      if exp_pol[i][j] == ag_pol[i][j]:
        count += 1

  acc = count/S_len
  print(acc)
  return acc
```

```python
def get_acc(pol, r_max, e):
  lambda_list = np.linspace(0, 5, 500)
  S_len = 100

  acc_list = []
  for l in lambda_list:
    c_1, D_1, b_1 = lp_matrices(l, pol, r_max)
    irl_reward = np.array(solve_irl(c_1, D_1, b_1)).squeeze()

    irl_reward = irl_reward.reshape(10,10)

    v_dis, v_hist_disc, step_disc, irl_opt = optimal(S, g, e, irl_reward, w)
    acc_list.append(accuracy(pol, irl_opt, S_len))
    print(l)

  plt.plot(lambda_list, acc_list)
  plt.xlabel("Lambda")
  plt.ylabel("Accuracy")
  plt.title("Accuracy as Lambda Increases")
  plt.show()

  return lambda_list, acc_list
```
